```python
import pandas as pd
```

```python
data = pd.read_csv('../Data/tableau_table.csv')
data.Q.value_counts()
```

```python
data.head(3)
```

```python

def probs(row,i):
    x = row[i]
    return float(x.split(',')[1].strip(']'))

prob_list = ['Q1_probabilities','Q2_probabilities','Q3_probabilities','Q4_probabilities']


for i,j in zip(prob_list,['Q1_Probs','Q2_Probs','Q3_Probs','Q4_Probs']):
    data[j] = data.apply( lambda row : probs(row, i), axis = 1)
    data
```

```python
data[prob_list+['Q1_Probs','Q2_Probs','Q3_Probs','Q4_Probs','Q']]
```

```python
    
X = data[['Q1_Probs','Q2_Probs','Q3_Probs','Q4_Probs']]
y = data.Q
X
```

```python
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV



x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = .2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=8, random_state=42)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
x_test['y_hat'] = y_pred
```

```python
x_test['true_y'] = y_test.to_list()
x_test[x_test.y_hat != x_test.true_y]
```

```python
z = len(x_test[x_test.y_hat != x_test.true_y])
zz = len(x_test)
```
